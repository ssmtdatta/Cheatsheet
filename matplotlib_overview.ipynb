### Representing data points.  

* marker symbols
* marker color (face, fill)
* marker size

* line color
* line thickness
* line type
* line plus marker

* x-label, y-label, title
* font style
* font color
* font size

* texts
* legends
* tick
* grid
* text/tick angle

```python
import numpy as np
import random

x = np.random.normal(size=100)
y = np.random.normal(size=100)

plt.plot(x, y, 'ro')
plt.show()
```

```python
x = np.random.normal(size=100)
y = np.random.normal(size=100)

plt.plot(x, y, 'gX')
plt.show()
```

```python
x = np.random.normal(size=100)
y = np.random.normal(size=100)

plt.plot(x, y, 'm3')
plt.show()
```

```python
#### grid
plt.grid(b=True, which='major', color='blue', linestyle = "-")
plt.grid(b=True, which='minor', color='black', linestyle='--')
```
